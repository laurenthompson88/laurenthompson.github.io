```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib import ticker
import numpy as np
```

```python
df = pd.read_excel('world_population.xlsm')
```

##  1 line chart

```python
fig, ax = plt.subplots(figsize=(10, 8))
plt.ylim(3e9, 7e9)
plt.plot(df['Year'], df['Population'])
plt.ylabel('Population in Billions', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.title('World Population from 1960 to 2009', fontsize=20)
ax.xaxis.set_major_locator(MultipleLocator(4))
ax.xaxis.set_major_formatter('{x:.0f}')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,pos: format(x/1000000000,'1.1f')+'B'))
# For the minor ticks, use no labels; default NullFormatter.
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(1e8))
plt.xlim(1958, 2011)
plt.grid()
plt.show()
```

## 1 step chart

```python

fig, ax = plt.subplots(figsize=(10, 8))
plt.ylim(3e9, 7e9)
plt.step(df['Year'], df['Population'], where='post', label='post')
plt.plot(df['Year'], df['Population'], 'o--', color='grey', alpha=0.3)
plt.ylabel('Population in Billions', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.title('World Population from 1960 to 2009', fontsize=20)
ax.xaxis.set_major_locator(MultipleLocator(4))
ax.xaxis.set_major_formatter('{x:.0f}')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x,pos: format(x/1000000000,'1.1f')+'B'))
# For the minor ticks, use no labels; default NullFormatter.
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(1e8))
plt.xlim(1958, 2011)
plt.grid()
plt.show()
```
