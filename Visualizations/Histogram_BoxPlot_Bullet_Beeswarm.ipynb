```python
# imports and read-ins
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.figure_factory as ff
import numpy as np


df = pd.read_csv('education.csv')
```

```python
df.head()
```

```python
df = df[df['state'] != 'United States']
```

## Histogram

```python
fig, ax = plt.subplots(figsize = (9, 9))
ax.hist(df['reading'], bins=15, edgecolor="black")
plt.xlabel('Reading Scores')
plt.title('Histogram of Reading Scores across the US')
```

## Box Plot

```python
plt.figure(figsize=(10, 10))
plt.xticks(rotation=90)
sns.boxplot(x=df['math'], y=df['reading'])
plt.xlabel('Math Scores')
plt.ylabel('Reading Scores')
plt.title('Math and Reading Scores across the US')
```

## Bullet Chart

```python
fig = go.Figure()

fig.add_trace(go.Indicator(
    mode = "number+gauge+delta",  value = int(df.loc[df['state']=='Alabama']['reading']),
    delta = {'reference': 501},
    domain = {'x': [0.25, 1], 'y': [0.08, 0.25]},
    title = {'text': "Alabama"},
    gauge = {
        'shape': "bullet",
        'axis': {'range': [None, 600]},
        'threshold': {
            'line': {'color': "black", 'width': 2},
            'thickness': 0.75,
            'value': 501}}))
fig.add_trace(go.Indicator(
    mode = "number+gauge+delta",  value = int(df.loc[df['state']=='Alaska']['reading']),
    delta = {'reference': 501},
    domain = {'x': [0.25, 1], 'y': [0.4, 0.6]},
    title = {'text': "Alaska"},
    gauge = {
        'shape': "bullet",
        'axis': {'range': [None, 600]},
        'threshold': {
            'line': {'color': "black", 'width': 2},
            'thickness': 0.75,
            'value': 501}}))
fig.add_trace(go.Indicator(
    mode = "number+gauge+delta",  value = int(df.loc[df['state']=='Arizona']['reading']),
    delta = {'reference': 501},
    domain = {'x': [0.25, 1], 'y': [0.7, 0.9]},
    title = {'text': "Arizona"},
    gauge = {
        'shape': "bullet",
        'axis': {'range': [None, 600]},
        'threshold': {
            'line': {'color': "black", 'width': 2},
            'thickness': 0.75,
            'value': 501}}))

fig.update_layout(height = 400 , margin = {'t':0, 'b':0, 'l':0})

fig.show()
```
