```python
# Set-up and imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.style as style
df = pd.read_excel('obama-approval-ratings.xls')
```

```python
#sets the size of the charts
style.use('seaborn-poster')
style.use('ggplot')
```

```python
df.head()
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Issue', y='Approve', data=df)
plt.title('Approved Count per Issue')
plt.xlabel('Issue')
plt.ylabel('Count')
plt.xticks(rotation=70)
plt.show()
```

```python
# Stacked Bar Chart
plt.bar(df['Issue'], df['Approve'])
plt.bar(df['Issue'], df['Disapprove'], bottom=df['Approve'])
plt.bar(df['Issue'], df['None'], bottom=df['Approve']+ df['Disapprove'])
plt.xlabel("Issues")
plt.ylabel("Count")
plt.legend(["Approve", "Disapprove", "None"],bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Approvals per Issue")
plt.xticks(rotation=70)
plt.show()
```

```python
# Pie Chart
fig, ax = plt.subplots()
ax.pie(df['Approve'], labels=df['Issue'], autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
ax.set_title('Approve Count per Issue')


plt.show()
```

```python
#Doughnut Chart
plt.pie(df['None'], labels=df['Issue'],
        autopct='%1.1f%%', pctdistance=0.85)
 
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
 
# Adding Title of chart
plt.title('None Votes per Issue')
 
# Displaying Chart
plt.show()
```
